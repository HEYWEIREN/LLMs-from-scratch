# 第 6 章练习答案（Chapter 6 Exercise solutions）

## 练习 6.1：增加 context length（Increasing the context length）

可以通过把 max length 设置为 1024，将 inputs padding 到模型支持的最大 token 数：

```python
max_length = 1024

train_dataset = SpamDataset(base_path / "train.csv", max_length=max_length, tokenizer=tokenizer)
val_dataset = SpamDataset(base_path / "validation.csv", max_length=max_length, tokenizer=tokenizer)
test_dataset = SpamDataset(base_path / "test.csv", max_length=max_length, tokenizer=tokenizer)
```

或者，也可以通过下面方式定义 `max_length`：

```python
max_length = model.pos_emb.weight.shape[0]
```

或者：

```python
max_length = BASE_CONFIG["context_length"]
```

为方便起见，可以用下面命令运行这个实验：

```bash
python additional-experiments.py --context_length "model_context_length"
```

使用 [../02_bonus_additional-experiments](../02_bonus_additional-experiments) 文件夹中的代码运行后，test accuracy 会明显更差，为 78.33%（主章节中是 95.67%）。

## 练习 6.2：微调整个模型（Finetuning the whole model）

如果不只是微调最后一个 transformer block，而是微调整个模型，可以从代码中移除下面几行：

```python
for param in model.parameters():
    param.requires_grad = False
```

为方便起见，可以用下面命令运行这个实验：

```bash
python additional-experiments.py --trainable_layers all
```

使用 [../02_bonus_additional-experiments](../02_bonus_additional-experiments) 文件夹中的代码运行后，test accuracy 提高约 1%，达到 96.67%（主章节中是 95.67%）。

## 练习 6.3：微调第一个 token 与最后一个 token（Finetuning the first versus last token）

如果不微调最后一个 output token，而是微调第一个 output token，可以把代码中所有：

```python
model(input_batch)[:, -1, :]
```

改成：

```python
model(input_batch)[:, 0, :]
```

为方便起见，可以用下面命令运行这个实验：

```
python additional-experiments.py --trainable_token first
```

使用 [../02_bonus_additional-experiments](../02_bonus_additional-experiments) 文件夹中的代码运行后，test accuracy 会明显更差，为 75.00%（主章节中是 95.67%）。
